## Project Summary: Quantum Counting for the Subset Sum Problem

### Project Overview
This project implements the **Quantum Counting Algorithm** to determine the number of valid solutions ($M$) for the **Subset Sum Problem**. In this problem, we seek to find how many subsets of a given list of weights sum exactly to a target value. While a classical computer requires $O(2^n)$ operations to check every combination, this quantum approach utilizes the relationship between **Grover’s Algorithm** and **Quantum Phase Estimation (QPE)** to estimate the count with a quadratic speedup.

### Methodology and Implementation
1.  **Constructive Phase Oracle**: We designed a quantum oracle that identifies "solution" states. It uses a series of **X-gates** and a **Multi-Controlled Z gate** (implemented via an `H-MCX-H` sandwich) to flip the phase of only those bitstrings whose weights match the target sum.
2.  **Grover Unitary ($G$)**: The oracle was combined with a **Diffusion Operator** to create a unitary rotation. The "speed" or angle of this rotation is directly proportional to the number of solutions in the search space ($N=2^n$).
3.  **Phase Estimation (QPE)**: We implemented a counting register ($t$ qubits) to measure the rotation's frequency. By applying **Controlled-Grover** iterations in powers of two ($G^{2^i}$), the phase information was "kicked back" into the counting register.
4.  **Inverse QFT & Measurement**: An **Inverse Quantum Fourier Transform** was used to convert the phase from the Fourier basis into a measurable binary integer.
5.  **Classical Post-Processing**: The measured integer was converted into a phase fraction ($\phi$), and the final solution count was calculated using the estimation formula:  
    $M = N \sin^2(\pi \phi)$

### Results
The implementation was validated using a **Precision Test Suite** across several scenarios:
*   **Accuracy**: For small $n$, the algorithm correctly identified counts for 0, 2, 3, and 6 solutions.
*   **Convergence**: By testing different values of $t$, we visualized how increasing the number of counting qubits significantly reduces the error margin, leading to a sharper "peak" in the measurement histogram.
*   **Symmetry Handling**: The code successfully managed the Grover symmetry ($M$ vs $N-M$) by assuming the solution set is the minority ($M < N/2$).

### Key Learnings
*   **Phase Kickback**: We gained a deep understanding of how information from a "Search" register can be encoded into a "Counting" register without measuring the search qubits directly.
*   **Geometric Rotation**: We learned to view Grover’s algorithm not just as a search, but as a precise geometric rotation in Hilbert space.
*   **Resource Trade-offs**: We observed the direct trade-off between the number of counting qubits (circuit width/depth) and the precision of the resulting estimate.
